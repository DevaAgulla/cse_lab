```python
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
```

```python
variety_counts = df["variety"].value_counts()
variety_counts = dict(variety_counts)
```

```python
a = variety_counts.keys()
b = variety_counts.values()
```

```python
import matplotlib.pyplot as plt
```

```python
#pie
fig = plt.figure(figsize=(9,9))
plt.pie(x = b,labels = a , explode=[0.1,0.1,0.1],shadow=True,autopct='%1.1f%%')
plt.title("iris Species %")
plt.show()
```

```python
import numpy as np
```

```python
fig = plt.figure(figsize=(7,7))
width = df["sepal.width"]
width = np.array([width[i] for i in range(0,50)])
length = df["sepal.length"]
length = np.array([length[i] for i in range(0,50)])
plt.xlim(4,8.5)
plt.ylim(1.5, 5)
plt.scatter(length,width , color = 'red')

width = df["sepal.width"]
width = np.array([width[i] for i in range(50,100)])
length = df["sepal.length"]
length = np.array([length[i] for i in range(50,100)])
plt.scatter(length,width, color = 'green')
width = df["sepal.width"]
width = np.array([width[i] for i in range(100,150)])
length = df["sepal.length"]
length = np.array([length[i] for i in range(100,150)])
plt.scatter(length,width ,color = 'blue')
plt.legend(["iris setosa","iris versicolor","iris virginica"])
plt.show()
```
