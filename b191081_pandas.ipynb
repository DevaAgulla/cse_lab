```python
import pandas as pd
```

```python
df = pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
print(df.shape) #no of rows and columns
```

```python
df = df.dropna()
```

```python
rooms = df["Rooms"]
```

```python
price = list(df["Price"])
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

plt.hist(x = price,edgecolor = "black",bins = 40)
plt.xlabel("price range")
plt.ylabel("NUmber of houses")
plt.title("Histogram of price")
plt.show()
```

```python
rooms_count = df["Rooms"].value_counts()
rooms_count = dict(rooms_count)
```

```python
a = rooms_count.keys()
b = rooms_count.values()
```

```python
#pie
fig = plt.figure()
plt.pie(x = b,labels = a)
plt.legend()
plt.show()
```

```python
region_names = dict(df["Regionname"].value_counts())
```

```python
a = region_names.keys()
b = region_names.values()
```

```python
#pie 
fig = plt.figure(figsize=(8,8))
plt.pie(x =b ,labels = a)
plt.legend()
plt.show()
```

```python
type_count = df["Type"].value_counts()
type_count = dict(type_count)
```

```python
a = list(type_count.keys())
b = list(type_count.values())
```

```python
#histogram
fig = plt.figure(figsize=(8,8))
plt.bar(x =a , height = b)
plt.show()
```

```python
#scatter plot
price = list(df["Price"])
distance = list(df["Distance"])
```

```python
fig = plt.figure(figsize = (10,10))
#line plot
plt.plot(distance,price,linestyle="--",marker="o",markerfacecolor="white", markeredgecolor="black")
plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance VS Price")
plt.show()
```
